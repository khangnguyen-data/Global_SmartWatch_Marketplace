# Part 3: Machine Learning

### Import Libaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('../Data/smartwatch_data_ready_for_ml.csv')
```

## 3.1: Training Machine Learning Model

### 3.1.1: Split data to X_train, X_valid, y_train, y_valid

```python
from sklearn.model_selection import train_test_split

X=data.drop(columns=['price', 'price_log'])
y=data['price_log']

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
X_valid.shape
```

```python
X.shape
```

### 3.1.2: Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

lr_model=LinearRegression()
lr_model.fit(X_train, y_train)
y_preds=lr_model.predict(X_valid)
```

```python
y_preds
```

```python
pd.DataFrame ({'y': y_valid.head(), 'y_pred': y_preds[:5]})
```

```python
y_valid_usd = np.expm1(y_valid)
y_pred_usd = np.expm1(y_preds)
lr_comparison_df = pd.DataFrame({
    'Actual Price ($)': y_valid_usd,
    'Predicted Price ($)': y_pred_usd,
    'Difference ($)': y_valid_usd - y_pred_usd
})

print(lr_comparison_df.head())
```

### 3.1.3: Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

rf_model=RandomForestRegressor(random_state=0)
rf_model.fit(X_train, y_train)
```

```python
rf_preds=rf_model.predict(X_valid)
```

```python
rf_preds[:5]
```

```python
pd.DataFrame ({'y': y_valid.head(), 'y_pred': rf_preds[:5]})
```

```python
rf_preds_usd = np.expm1(rf_preds)
rf_comparison_df = pd.DataFrame({
    'Actual Price ($)': y_valid_usd,
    'Predicted Price ($)': rf_preds_usd,
    'Difference ($)': y_valid_usd - rf_preds_usd
})

print(rf_comparison_df.head())
```

### 3.1.4: XGBoost Model

```python
import xgboost as xgb
xgb_model = xgb.XGBRegressor(random_state=0)
xgb_model.fit(X_train, y_train)
xgb_preds = xgb_model.predict(X_valid)

xgb_preds[:5]
pd.DataFrame({'y': y_valid.head(), 'y_pred': xgb_preds[:5]})

xgb_preds_usd = np.expm1(xgb_preds)
xgb_comparison_df = pd.DataFrame({
    'Actual Price ($)': y_valid_usd,
    'Predicted Price ($)': xgb_preds_usd,
    'Difference ($)': y_valid_usd - xgb_preds_usd
})
print(xgb_comparison_df.head())
```

## 3.2: Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def get_metrics(y_true_log, y_pred_log):
    y_true = np.expm1(y_true_log)
    y_pred = np.expm1(y_pred_log)
    return {
        'MAE': mean_absolute_error(y_true, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_true, y_pred)),
        'R2': r2_score(y_true, y_pred)
    }
results = pd.DataFrame({
    'Linear Regression': get_metrics(y_valid, y_preds),
    'Random Forest': get_metrics(y_valid, rf_preds),
    'XGBoost': get_metrics(y_valid, xgb_preds)
}).T

print(results)
```

#### Ta thấy XGBoost Model tốt hơn so với 2 Model còn lại:
* XGBoost có tỉ lệ 0.422 nghĩa là Model đã hiểu được khoảng 42.2% lý do tại sao giá đồng hồ lại tăng hay giảm trong khi Linear Regression chỉ có 18,7%
* Con số 42.2% còn quá thấp nên ta quyết định sử dụng XGBoost Model và tìm cách tằn R2 lên

#### Giải thích các thuật ngữ:
* MAE: Sai số tuyệt đối trung bình
* RMSE: Căn bậc hai sai số bình phương trung bình
* R2: Hệ số xác định

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200],           
    'max_depth': [3, 6],                  
    'learning_rate': [0.05, 0.1],         
    'subsample': [0.8, 1.0]  
}             
grid_search = GridSearchCV(
    estimator=xgb.XGBRegressor(random_state=42),
    param_grid=param_grid,
    cv=5,                
    scoring='r2',        
    n_jobs=-1,            
    verbose=1            
)
grid_search.fit(X_train, y_train)

print(f"Cấu hình tốt nhất: {grid_search.best_params_}")
print(f"R2 cao nhất đạt được: {grid_search.best_score_:.4f}")

best_xgb = grid_search.best_estimator_
final_preds_log = best_xgb.predict(X_valid)
```
